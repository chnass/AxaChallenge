```python
print(f"Number of accidents recorded: {len(df)}")
```

```python
df_bike = df[df[[f"VEHICLE TYPE CODE {x}" for x in range(1,6)]].apply(lambda col: col.astype(str).str.contains('Bike').any(), axis=1)]
```

```python
print(f"In {len(df_bike)/len(df)*100:.2f}% of the accidents a Bike or E-Bike is involved.")
```

```python
print(f"{len(df_bike)} crashes with a Bike or E-Bike were recorded between {df_bike["DATE"].min()} and {df_bike["DATE"].max()}")
```

```python
df_rbike = df_bike[df_bike[[f"VEHICLE TYPE CODE {x}" for x in range(1,6)]].apply(lambda col: (col.astype(str)=="Bike").any(), axis=1)]
df_ebike = df_bike[df_bike[[f"VEHICLE TYPE CODE {x}" for x in range(1,6)]].apply(lambda col: (col.astype(str)=="E-Bike").any(), axis=1)]
print(f"{len(df_rbike)} ({len(df_rbike)/len(df_bike)*100:.2f}%) of them were with a regular (non-electric) bike")
```

```python
# New reporting system fully enroled in 2016
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 7))
plt.subplots_adjust(bottom=0.0, right=0.9, top=1.0)
df_bike["DATE"].groupby(df_bike["DATE"].dt.year).count().plot(ax=axs[0,0], kind="bar", xlabel="Year", ylabel="Number of crashes")
df_bike["DATE"].groupby(df_bike["DATE"].dt.month).count().plot(ax=axs[0,1], kind="bar", xlabel="Month", ylabel="Number of crashes")
ax = df_bike["DATE"].groupby(df_bike["DATE"].dt.day_of_week).count().plot(ax=axs[1,0], kind="bar", xlabel="", ylabel="Number of crashes")
ax.set_xticks([0,1,2,3,4,5,6], labels=["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
df_bike["DATE"].groupby(df_bike["DATE"].dt.hour).count().plot(ax=axs[1,1], kind="bar", xlabel="Hour", ylabel="Number of crashes")
# df_bike["BOROUGH"].groupby(df_bike["BOROUGH"]).count().plot(ax=axs[1,1], kind="bar", ylabel="Number of crashes")
```
